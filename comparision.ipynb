Mini Project - Text File Conversion
---
Compare the performance (time taken) of C, C++, Java, R, and Python programs for converting 200MB, 400MB, 600MB, 800MB, and 1000MB text files to **upper-case.**

**Sample Result**

| File Size |  C  | C++ | Java |  R  | Python |
| --------- | --- | --- | ---- | --- | ------ |
|  200MB    | 12s | 15s |  18s | 20s |   25s  |
|  400MB    | 20s | 25s |  30s | 35s |   40s  |
|  600MB    | 34s | 36s |  40s | 45s |   53s  |
|  800MB    | 45s | 50s |  55s | 60s |   75s  |
| 1000MB    | 55s | 60s |  70s | 80s |  100s  |

```python
import matplotlib.pyplot as plt

file_sizes = (200, 400, 600, 800, 1000)
c = (12, 20, 34, 45, 55)
cpp = (15, 25, 36, 50, 60)
java = (18, 30, 40, 55, 70)
r = (20, 35, 45, 60, 80)
py = (25, 40, 53, 75, 100)

plt.title("Performance Comparision")
plt.xlabel("File Size (MB)")
plt.ylabel("Time Taken (sec)")

plt.plot(file_sizes, c, 'bD-', label="C")
plt.plot(file_sizes, cpp, 'rs-', label="C++")
plt.plot(file_sizes, java, 'g^-', label="Java")
plt.plot(file_sizes, r, 'mx-', label="R")
plt.plot(file_sizes, py, 'c2-', label="Python")

plt.legend()
plt.show()
```
