# Mini Project - Text File Conversion
---
Compare the performance (time taken) of C, C++, Java, R, and Python programs for converting 200MB, 400MB, 600MB, 800MB, and 1000MB text files to **upper-case.**

## Programs to convert text to upper case

### 1. C

```c
#include <stdio.h>
#include <ctype.h>

int main() {
    FILE *infile = fopen("input.txt", "r");
    FILE *outfile = fopen("output.txt", "w");

    if (infile == NULL || outfile == NULL) {
        printf("Error opening file.\n");
        return 1;
    }

    int ch;
    while ((ch = fgetc(infile)) != EOF) {
        fputc(toupper(ch), outfile);
    }

    fclose(infile);
    fclose(outfile);
    return 0;
}
```

### 2. C++

```cpp
#include <iostream>
#include <fstream>
#include <cctype>

int main() {
    std::ifstream infile("input.txt");
    std::ofstream outfile("output.txt");

    if (!infile || !outfile) {
        std::cerr << "File error.\n";
        return 1;
    }

    char ch;
    while (infile.get(ch)) {
        outfile.put(std::toupper(ch));
    }

    infile.close();
    outfile.close();
    return 0;
}
```

### 3. Java

```java
import java.io.*;

public class UpperCaseConverter {
    public static void main(String[] args) {
        try (
            BufferedReader reader = new BufferedReader(new FileReader("input.txt"));
            BufferedWriter writer = new BufferedWriter(new FileWriter("output.txt"));
        ) {
            int ch;
            while ((ch = reader.read()) != -1) {
                writer.write(Character.toUpperCase(ch));
            }
        } catch (IOException e) {
            System.out.println("File error: " + e.getMessage());
        }
    }
}

```

### 4. R

```r

```

## **Results**

| File Size |  C  | C++ | Java |  R  | Python |
| --------- | --- | --- | ---- | --- | ------ |
|  200MB    | 11s | 14s |  17s | 22s |   27s  |
|  400MB    | 19s | 23s |  29s | 37s |   44s  |
|  600MB    | 32s | 35s |  42s | 50s |   58s  |
|  800MB    | 44s | 48s |  57s | 65s |   82s  |
| 1000MB    | 54s | 59s |  72s | 85s |  107s  |

````python
import matplotlib.pyplot as plt

file_sizes = (200, 400, 600, 800, 1000)
c    = (11, 19, 32, 44,  54)
cpp  = (14, 23, 35, 48,  59)
java = (17, 29, 42, 57,  72)
r    = (22, 37, 50, 65,  85)
py   = (27, 44, 58, 82, 107)

plt.title("Performance Comparision")
plt.xlabel("File Size (MB)")
plt.ylabel("Time Taken (sec)")

plt.plot(file_sizes, c, 'bD-', label="C")
plt.plot(file_sizes, cpp, 'rs-', label="C++")
plt.plot(file_sizes, java, 'g^-', label="Java")
plt.plot(file_sizes, r, 'mx-', label="R")
plt.plot(file_sizes, py, 'c2-', label="Python")

plt.legend()
plt.show()
````

---
**Observations:-**
1. C is the fastest, while Python is the slowest
2. Lower-level languages like C and C++ are substantially more faster than the higher-level languages like Java, R, and Python.
3. The code in C is the most verbose, while Python has the smallest and easiest to understand code.


**Conclusion:-**
As the ease in understanding of a programming language increases, it's execution speed appears to decrease.
